```python
ori_img = os.path.join(root_dir, 'images')
```

```python
dst_root_dir = ori_img.replace('CAT2000', 'nCAT2000')
if not os.path.exists(dst_root_dir): os.makedirs(dst_root_dir)
```

```python
cate = [item for item in os.listdir(ori_img) if not item.startswith('.')]
cate
```

```python
cnt = 0
for tmp_cate in cate:
    tmp_png_dir = os.path.join(ori_img, tmp_cate)
    
#     if not os.path.exists(tmp_png_dir):
#         os.makedirs(tmp_png_dir)
    
    # print(tmp_png_dir)
        
    fixs = os.listdir(tmp_png_dir)
    
    # print(f'{tmp_cate} has {len(fixs)}')
    cnt += len(fixs)
    
    for fix in fixs:
        cur_dir = os.path.join(tmp_png_dir, fix)
        img_name = '_'.join(cur_dir.split('/')[-2:])
        cur_dst_dir = os.path.join(dst_root_dir, img_name)
        shutil.copy2(cur_dir, cur_dst_dir)
        print(cur_dir, cur_dst_dir)
```
